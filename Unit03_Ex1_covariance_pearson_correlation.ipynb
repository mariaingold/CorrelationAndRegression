# Pearson's Correlation between two variables

## Covariance

- Covariance: directional relationship between 2 variables
- Positive covariance: increase or decrease together
- Negative covariance: Move in opposite directions

Affected by
- Shifts to mean (add or subtract) do not affect covariance
- Increasing standard deviation tends to increase magnitude
- Larger sd means more spread in data

## Pearson's Correlation

- Standardised measure of linear releationship between 2 variables
- Ranges from -1 (perfect negative correlation) to +1 (perfect positive correlation)
- 0 means no linear relationship

Affected by

- Shifts to mean (add or subtract) do not affect covariance
- Not affected by standard deviation

```python

# calculate the Pearson's correlation between two variables
from numpy import mean
from numpy import std
from numpy import cov
from numpy.random import randn
from numpy.random import seed
from matplotlib import pyplot as plt
import seaborn as sns

from scipy.stats import pearsonr

# seed random number generator
#
# 1: Longer and thinner. Slightly higher. One outlier far top right.
# data1: mean=100.776 stdv=19.620
# data2: mean=151.050 stdv=22.358
# Covariance: 389.755
# Pearsons correlation: 0.888
#
# 42: Wider. Lower. One outlier far top right.
# data1: mean=100.387 stdv=19.575
# data2: mean=151.095 stdv=21.605
# Covariance: 375.654
# Pearsons correlation: 0.887
seed(1)

# prepare data
data1 = 20 * randn(1000) + 100
data2 = data1 + (10 * randn(1000) + 50)

# calculate covariance matrix
covariance = cov(data1, data2)

# calculate Pearson's correlation
corr, _ = pearsonr(data1, data2)

# plot
plt.scatter(data1, data2)
plt.show()

# summarize
print('data1: mean=%.3f stdv=%.3f' % (mean(data1), std(data1)))
print('data2: mean=%.3f stdv=%.3f' % (mean(data2), std(data2)))
print('Covariance: %.3f' % covariance[0][1])
print('Pearsons correlation: %.3f' % corr)
```
